```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.artist import Artist  
```

# Finitely Optimal

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
plt.rcParams['font.size'] = '14'

# Draw constraint lines

ax.plot(np.linspace(-1, 14, 100), np.linspace(6.5, -1, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(3.5, 11, 100))
ax.plot(np.linspace(10, 10, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(0, 0, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(0, 0, 100))
# ax.text(3, 8, "$x_1 - 2x_2 \leq -8$")
# ax.text(11, 1, "$x_1 + 2x_2 \leq 12$")
# ax.text(10.5, 5, "$x_1 \leq 10$")
# ax.text(-2, 2, "$x_1 \geq 0$")
# ax.text(4, -1, "$x_2 \geq 0$")


ax.set_xlim(-3, 15)

for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
    
ax.set_xlabel("x1")
ax.set_ylabel("x2")

# Draw the feasible region
feasible_set = Polygon(np.array([[0, 0], 
                                 [10, 0], 
                                 [10, 1], 
                                 [2, 5],
                                 [0,4]]))
ax.add_artist(feasible_set)
Artist.set_alpha(feasible_set, 0.2) 

# Draw the objective function
ax.plot(np.linspace(8, 12, 100),  np.linspace(5, -3, 100), 'k-')
ax.text(11, 3, "$2x_1 + x_2 = 21$")

ax.plot(10, 1, "o", color="black")
ax.text(11, 2, "optimal solution")
```

# Not Feasible 

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
plt.rcParams['font.size'] = '14'

# Draw constraint lines

ax.plot(np.linspace(-1, 12, 100), np.linspace(4.5, -2, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(4.5, 12, 100))
ax.plot(np.linspace(10, 10, 100), np.linspace(-2, 12, 100))
ax.plot(np.linspace(0, 0, 100), np.linspace(-2, 12, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(0, 0, 100))
ax.text(1.5, 5, "$x_1 - 2x_2 \leq -10$")
ax.text(6, 2, "$x_1 + 2x_2 \leq 8$")
ax.text(10.5, 5, "$x_1 \leq 10$")
ax.text(-2, 2.5, "$x_1 \geq 0$")
ax.text(4, -1, "$x_2 \geq 0$")

ax.set_xlim(-3, 15)

for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
    
ax.set_xlabel("x1")
ax.set_ylabel("x2")
```

# Unbounded

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
plt.rcParams['font.size'] = '14'

# Draw constraint lines

ax.plot(np.linspace(-1, 4, 100), np.linspace(8, -2, 100))
ax.plot(np.linspace(-1, 8, 100), np.linspace(3, -1, 100))
ax.plot(np.linspace(0, 0, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(-1, 12, 100), np.linspace(0, 0, 100))
ax.text(4.5, -2, "$2x_1 - x_2 \geq 6$")
ax.text(8.5, -1, "$x_1 + 2x_2 \geq 6$")
ax.text(-2, 1, "$x_1 \geq 0$")
ax.text(1, -1, "$x_2 \geq 0$")

ax.set_xlim(-3, 13)

for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
    
ax.set_xlabel("x1")
ax.set_ylabel("x2")

# Draw the feasible region
feasible_set = Polygon(np.array([[12, 0], 
                                 [12, 10], 
                                 [0, 10], 
                                 [0, 6],
                                 [2.2,1.5],
                                 [6,0]]))
ax.add_artist(feasible_set)
Artist.set_alpha(feasible_set, 0.2) 
```
