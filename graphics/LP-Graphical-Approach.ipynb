# 1. Draw equations  

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
plt.rcParams['font.size'] = '14'

# Draw constraint lines

ax.plot(np.linspace(-1, 14, 100), np.linspace(6.5, -1, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(3.5, 11, 100))
ax.plot(np.linspace(10, 10, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(0, 0, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(0, 0, 100))
ax.text(3, 8, "$x_1 - 2x_2 = -8$")
ax.text(11, 1, "$x_1 + 2x_2 = 12$")
ax.text(10.5, 5, "$x_1 = 10$")
ax.text(-2, 2, "$x_1 = 0$")
ax.text(4, -1, "$x_2 = 0$")
# ax.text(1.5, 8, "$x_1 - 2x_2 \geq -8$")
# ax.text(11, 1.5, "$x_1 + 2x_2 \leq 12$")
# ax.text(10.5, 5, "$x_1 \leq 10$")
# ax.text(-2, 2.5, "$x_1 \geq 0$")
# ax.text(4, -1, "$x_2 \geq 0$")

ax.set_xlim(-3, 15)

for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
    
ax.set_xlabel("x1")
ax.set_ylabel("x2")
```

# 2. Draw Feasible Region

## 2.1 Draw each region individually 

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
plt.rcParams['font.size'] = '14'

# Draw constraint lines

ax.plot(np.linspace(-1, 14, 100), np.linspace(6.5, -1, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(3.5, 11, 100))
ax.plot(np.linspace(10, 10, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(0, 0, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(0, 0, 100))
ax.text(3, 8, "$x_1 - 2x_2 \geq -8$")
# ax.text(11, 1, "$x_1 + 2x_2 \leq 12$")
# ax.text(10.5, 5, "$x_1 \leq 10$")
# ax.text(-2, 2, "$x_1 \geq 0$")
# ax.text(4, -1, "$x_2 \geq 0$")

ax.set_xlim(-3, 15)

for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
    
ax.set_xlabel("x1")
ax.set_ylabel("x2")

# Draw the feasible region
feasible_set_1 = Polygon(
    np.array(
        [
            [0, -1], 
            [14, -1], 
            [14, 10], 
            [0, 10],
        ]
    )
)
ax.add_artist(feasible_set_1)
Artist.set_alpha(feasible_set_1, 0.2) 
Artist.set_visible(feasible_set_1, False)

feasible_set_2 = Polygon(
    np.array(
        [
            [-1, 0], 
            [14, 0], 
            [14, 10], 
            [-1,10]
        ]
    )
)
ax.add_artist(feasible_set_2)
Artist.set_alpha(feasible_set_2, 0.2)
Artist.set_visible(feasible_set_2, False)

feasible_set_3 = Polygon(
    np.array(
        [
            [10, -1], 
            [10, 10], 
            [-1, 10], 
            [-1,-1]
        ]
    )
)
ax.add_artist(feasible_set_3)
Artist.set_alpha(feasible_set_3, 0.2)
Artist.set_visible(feasible_set_3, False)

feasible_set_4 = Polygon(
    np.array(
        [
            [-1, 3.5], 
            [14, 11], 
            [14, -1], 
            [-1,-1]
        ]
    )
)
ax.add_artist(feasible_set_4)
Artist.set_alpha(feasible_set_4, 0.2)
Artist.set_visible(feasible_set_4, True)

feasible_set_5 = Polygon(
    np.array(
        [
            [-1, 6.5], 
            [-1, -1], 
            [14, -1], 
        ]
    )
)
ax.add_artist(feasible_set_5)
Artist.set_alpha(feasible_set_5, 0.2)
Artist.set_visible(feasible_set_5, False)
```

## 2.2 Draw intersecting region 

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
plt.rcParams['font.size'] = '14'

# Draw constraint lines

ax.plot(np.linspace(-1, 14, 100), np.linspace(6.5, -1, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(3.5, 11, 100))
ax.plot(np.linspace(10, 10, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(0, 0, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(0, 0, 100))
ax.text(3, 8, "$x_1 - 2x_2 \geq -8$")
ax.text(11, 1, "$x_1 + 2x_2 \leq 12$")
ax.text(10.5, 5, "$x_1 \leq 10$")
ax.text(-2, 2, "$x_1 \geq 0$")
ax.text(4, -1, "$x_2 \geq 0$")

ax.set_xlim(-3, 15)

for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
    
ax.set_xlabel("x1")
ax.set_ylabel("x2")

# Draw the feasible region
feasible_set = Polygon(np.array([[0, 0], 
                                 [10, 0], 
                                 [10, 1], 
                                 [2, 5],
                                 [0,4]]))
ax.add_artist(feasible_set)
Artist.set_alpha(feasible_set, 0.2) 
```

# Draw isoquant lines

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
plt.rcParams['font.size'] = '14'

# Draw constraint lines

ax.plot(np.linspace(-1, 14, 100), np.linspace(6.5, -1, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(3.5, 11, 100))
ax.plot(np.linspace(10, 10, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(0, 0, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(0, 0, 100))

ax.set_xlim(-3, 15)

for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
    
ax.set_xlabel("x1")
ax.set_ylabel("x2")

# Draw the feasible region
feasible_set = Polygon(np.array([[0, 0], 
                                 [10, 0], 
                                 [10, 1], 
                                 [2, 5],
                                 [0,4]]))
ax.add_artist(feasible_set)
Artist.set_alpha(feasible_set, 0.2) 

# Draw the objective function
ax.plot(np.linspace(-1, 3, 100),  np.linspace(6, -2, 100), 'k-')
ax.plot(np.linspace(-1, 4, 100),  np.linspace(8, -2, 100), 'k-')
ax.plot(np.linspace(-1, 5, 100), np.linspace(10, -2, 100), 'k-')
ax.text(1.5, 8, "$2x_1 + x_2 = 8$")
ax.text(1.25, 9, "$2x_1 + x_2 = 6$")
ax.text(1.0, 10, "$2x_1 + x_2 = 4$")
```

# 4. Move isoqunat lines until optimal solution

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
plt.rcParams['font.size'] = '14'

# Draw constraint lines

ax.plot(np.linspace(-1, 14, 100), np.linspace(6.5, -1, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(3.5, 11, 100))
ax.plot(np.linspace(10, 10, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(0, 0, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(0, 0, 100))

ax.set_xlim(-3, 15)

for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
    
ax.set_xlabel("x1")
ax.set_ylabel("x2")

# Draw the feasible region
feasible_set = Polygon(np.array([[0, 0], 
                                 [10, 0], 
                                 [10, 1], 
                                 [2, 5],
                                 [0,4]]))
ax.add_artist(feasible_set)
Artist.set_alpha(feasible_set, 0.2) 

# Draw the objective function
ax.plot(np.linspace(8, 12, 100),  np.linspace(5, -3, 100), 'k-')
ax.text(11, 3, "$2x_1 + x_2 = 21$")

ax.plot(10, 1, "o", color="black")
ax.text(11, 2, "optimal solution")
```

# 5. Find binding constraints 

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
plt.rcParams['font.size'] = '14'

# Draw constraint lines

ax.plot(np.linspace(-1, 14, 100), np.linspace(6.5, -1, 100), linewidth=4)
ax.plot(np.linspace(-1, 14, 100), np.linspace(3.5, 11, 100))
ax.plot(np.linspace(10, 10, 100), np.linspace(-1, 10, 100), linewidth=4)
ax.plot(np.linspace(0, 0, 100), np.linspace(-1, 10, 100))
ax.plot(np.linspace(-1, 14, 100), np.linspace(0, 0, 100))
ax.text(5 ,5, "$x_1 + 2x_2 \leq 12$")
ax.text(10.5, 6, "$x_1 \leq 10$")
ax.text(5 ,4, "binding")
ax.text(10.5, 5, "binding")

ax.set_xlim(-3, 15)

for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')
    
ax.set_xlabel("x1")
ax.set_ylabel("x2")

# Draw the feasible region
feasible_set = Polygon(np.array([[0, 0], 
                                 [10, 0], 
                                 [10, 1], 
                                 [2, 5],
                                 [0,4]]))
ax.add_artist(feasible_set)
Artist.set_alpha(feasible_set, 0.2) 

# Draw the objective function
# ax.plot(np.linspace(8, 12, 100),  np.linspace(5, -3, 100), 'k-')
# ax.text(11, 3, "$2x_1 + x_2 = 21$")

ax.plot(10, 1, "o", color="black")
ax.text(10.5, 1.5, "optimal solution")
```
